- What is the expected value of a single die roll?

#### Answer
Since a die roll has discrete outcomes of either {1, 2, 3, 4, 5, 6}, the expected value can be expressed as:

$\mathbb{E}[X] = \sum_{x \in \text{supp}(X)} x \times m_X(x)$

For a fair six sided die, $m(x) = \frac{1}{6}$ for each $x$

Thus, the expected value of a single die roll would be:

$(1)(\frac{1}{6}) + (2)(\frac{1}{6}) + (3)(\frac{1}{6}) + (4)(\frac{1}{6}) + (5)(\frac{1}{6}) + (6)(\frac{1}{6}) = \frac{21}{6} = 3.5$

The expected value of a single die roll is $\bold{3.5}$

- What is the expected value of rolling two dice and adding the results together?

#### Answer
Each die has 6 outcomes, so there are 6 * 6 = 36 total outcomes for the roll. The possible sums range from 2 to 12, and some sums can result from more combinations as opposed to other sums. For example, two can only occur from one outcome, but 7 can occur from 6 outcomes of the roll. Counting this for each possible sum, we get the following expected value:

$\mathbb{E}[X] = (2)(\frac{1}{36}) + (3)(\frac{2}{36}) + (4)(\frac{3}{36}) + (5)(\frac{4}{36}) + (6)(\frac{5}{36}) + (7)(\frac{6}{36}) + (8)(\frac{5}{36}) + (9)(\frac{4}{36}) + (10)(\frac{3}{36}) + (11)(\frac{2}{36}) + (12)(\frac{1}{36}) = (\frac{252}{36}) = 7$

Thus the expected value of rolling two dice and adding the results together is $\bold{7}$

- What is the expected winnings of any gamble in European roulette?

#### Answer
In a gamble on European roulette, assume you pay 1 dollar to bet on $K$ slots out of 37 total slots. Thus, you have a probability of $K/37$ to win and get $36/K$ dollars, but you have a $37-K/37$ chance of loosing and getting nothing. Therefore the expected payoff is:

$\mathbb{E}[X] = \frac{K}{37} * \frac{36}{K} + \frac{37-K}{37} * 0 - 1 = \frac{36}{37} - 1 = -0.027$

Thus you are expected to lose about 2.7 cents for every dollar bet on European roulette.

- Imagine that the process described in the last question continues until you fail to roll a six. What is the expected value of the process? (This can be tricky, you can simulate it to get an answer if you prefer. Hint: The answer is 4.2.)

```python
# Simulate above process until we fail to roll a 6
import numpy as np

rng = np.random.default_rng(123)

def roll_die():
  sum = 0
  play = 1
  while play == 1:
    roll = rng.integers(1, 7)
    sum += roll
    if roll != 6:
      play = 0
  
  return sum

n_sim = 1000000
results = [roll_die() for i in range(n_sim)]

print(np.mean(results))
```

From the above simulation code, we get a value of 4.196006, which is very close to the correct answer of 4.2. So the expected value is 4.2

- Compute the expected value for a uniform random variable.

#### Answer
For a uniform random variable with supp($X$) = [$a, b$] and $f_X(x) = \frac{1}{b-a}$, we get the following expectation formula to solve:

$\mathbb{E}[X] = \int_{a}^{b} x \cdot \frac{1}{b-a} \, dx = \frac{1}{b-a} \cdot \int_{a}^{b} x \, dx = $

$\frac{1}{b-a} \cdot \frac{b^2 - a^2}{2} = \bold{\frac{b+a}{2}}$

Thus, the expected value of a uniform random variable is exactly the midpoint of the interval/support for that variable. For example, if we have values ranging from 0 to 1, the expected value would be 1/2

- The **covariance** of $X$ and $Y$ is
$$
\text{cov}(X,Y) = \int_{y} \int_{x} (x-\mathbb{E}[X])(y-\mathbb{E}[Y])f_{XY}(x,y) dxdy = \mathbb{E}_{XY}[ (x-\mu_X)(y-\mu_Y)]
$$
- Show that if $f_{XY}(x,y)=f_X(x)f_Y(y)$, then $\text{cov}(X,Y)=0$
#### Answer
Substituting the above into the covariance equation we start with the following:
\begin{alignat*}{2}
\text{cov}(X,Y) &=& \int_{y} \int_{x} (x-\mathbb{E}[X])(y-\mathbb{E}[Y])f_{X}(x)f_y(y) dxdy \\
&=& \int_{y} \left[ \int_{x} (x-\mathbb{E}[X])f_{X}(x) dx \right] (y-\mathbb{E}[Y])F_y(y)dy \\
&=& \int_{y} \left[ \int_{x} xf_{X}(x) dx-\mathbb{E}[X] \right] (y-\mathbb{E}[Y])F_y(y)dy \\
&=& \int_{y} \left[ \mathbb{E}[X]-\mathbb{E}[X] \right] (y-\mathbb{E}[Y])F_y(y)dy \\
&=& \int_{y} \left[ 0 \right] (y-\mathbb{E}[Y])F_y(y)dy \\
&=& 0
\end{alignat*}

As seen above, after evaluating the X integral we see it results in E[x] - E[x] or simply 0. Thus, we have shown that if $f_{XY}(x,y)=f_X(x)f_Y(y)$, then $\text{cov}(X,Y)=0$

Suppose $X$ has an expectation $\mathbb{E}[X]<\infty$ and variance $\mathbb{V}[X]<\infty$; this isn't always true, but is *usually* true
- Consider making a new variable, $\varepsilon = X - \mathbb{E}[X]$
- What's the expectation of $\varepsilon$?

#### Answer
The expectation of $\varepsilon$ can be determined as follows:

\begin{alignat*}{2}
\mathbb{E}[\varepsilon] &=& \int (x - \mathbb{E}[X]) f(x) dx \\
 &=& \int xf(x)dx - \mathbb{E}[X] \\
 &=& \mathbb{E}[X] - \mathbb{E}[X] \\
 &=& 0
\end{alignat*}

Thus, $\mathbb{E}[\varepsilon] = 0$


- What's the variance of $\varepsilon$?

#### Answer
The variance of $\varepsilon$ can be determined as follows:

\begin{alignat*}{2}
\mathbb{V}[\varepsilon] &=& \int (x - \mathbb{E}[X] - \mathbb{E}[x - \mathbb{E}[X]])^2 f(x) dx \\
 &=& \int (x - \mathbb{E}[X])^2 f(x) dx \\
  &=& \mathbb{V}[X]
\end{alignat*}

Thus $\mathbb{V}[\varepsilon] = \mathbb{V}[X]$
