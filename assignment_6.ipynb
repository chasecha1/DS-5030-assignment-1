2.1 What are the outcomes for rolling a single, fair **THREE**-sided die? What's the set of all events? What are the probabilities of all the events?

#### Answer
Assuming we roll the die only once, the possible outcomes would be rolling a 1, 2, or a 3. Thus, the set of all outcomes can be represented by $Z = \{1, 2, 3\}$ which is our sample space

The total number of events is $2^3 = 8$. An event $E$ is any subset of the outcomes, and thus the set of all events is simply the powerset of our sample space, or $\mathcal{P}(Z)$. This set written out would be $E = \{\emptyset, \{1\}, \{2\}, \{3\}, \{1, 2\}, \{1, 3\}, \{2, 3\}, \{1, 2, 3\}  \}$

As for the probability of all the events, we can assume that the die is fair and that each outcome is equally likely meaning that $P(\{1\}) = P(\{2\}) = P(\{3\}) = 1/3$. Then, every event is simply the sum of the probabilites of the outcomes it contains since the outcomes are disjoint and their probabilities sum to $1$. Written out, we get the following:

$P(\{\emptyset\}) = 0$: This is an impossible event as we must roll some number when rolling the three sided die.

$P(\{1\}) = P(\{2\}) = P(\{3\}) = 1/3$

$P(\{1, 2\}) = P(\{1, 3\}) = P(\{2, 3\}) = 2/3$

$P(\{1, 2, 3\}) = 1$: This is a certain event because one of the outcomes is guaranteed to occur.

- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.


#### Answer
- Support: The support of the logistic distribution is all real numbers or $x \in (-\infty, \infty)$
- Density
    - $f(x) = \dfrac{e^{\frac{-x}{\sigma}}}{\sigma(e^{\frac{-x}{\sigma}}+1)^2}$
- Plotting density and distribution for variety of sigma:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp

def logistic_cdf(x, sigma):
    return 1 / (1 + np.exp(-x / sigma))

def logistic_pdf(x, sigma):
    ex = np.exp(-x / sigma)
    return (1 / sigma) * ex / (1 + ex)**2

x_log = np.linspace(-10, 10, 1001)

sigmas = [0.5, 1.0, 2.0]

# Logistic CDFs
for sigma in sigmas:
    plt.figure()
    plt.plot(x_log, logistic_cdf(x_log, sigma), label=f'σ={sigma}')
    plt.title('Logistic Distribution')
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.legend()
    plt.grid(True)

# Logistic PDFs
for sigma in sigmas:
    plt.figure()
    plt.plot(x_log, logistic_pdf(x_log, sigma), label=f'σ={sigma}')
    plt.title(f'Logistic Density')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
```

```python
# Sample with 1000 draws, plot KDE/ECDF compared to theoretical pdf/cdf

grid = np.linspace(-10, 10, 1001)
sample = sp.stats.logistic.rvs(size=1000, random_state=10)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
sns.lineplot(x=grid,y=logistic_pdf(grid, 1),ax=axes[0],label='pdf',linestyle='--')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=logistic_cdf(grid, 1),ax=axes[1],label='cdf',linestyle='--')
axes[1].set_title('Sample ECDF')
plt.show()
```

- Exponential distribution (similar to log-normal): (use suvrival in months)
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.

#### Answer
- Support: The support of the exponential distribution is from 0 to infinity, or $x \in [0, \infty)$
- Density:
    - $f(x) = \lambda e^{-\lambda x}$
- Plotting density and distribution for variety of lambda:

```python
def exp_cdf(x, lam):
    return np.where(x >= 0, 1 - np.exp(-lam * x), 0)

def exp_pdf(x, lam):
    return np.where(x >= 0, lam * np.exp(-lam * x), 0)

x_exp = np.linspace(0, 10, 1001)
lambdas = [0.5, 1.0, 2.0]

# Exponential CDFs
for lam in lambdas:
    plt.figure()
    plt.plot(x_exp, exp_cdf(x_exp, lam), label=f'λ={lam}')
    plt.title('Exponential Distribution')
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.legend()
    plt.grid(True)

# Exponential PDFs
for lam in lambdas:
    plt.figure()
    plt.plot(x_exp, exp_pdf(x_exp, lam), label=f'λ={lam}')
    plt.title(f'Exponential Density')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
```

```python
# Sample with 1000 draws, plot KDE/ECDF compared to theoretical pdf/cdf

grid = np.linspace(0, 10, 1001)
sample = sp.stats.expon.rvs(size=1000, random_state=10)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
sns.lineplot(x=grid,y=exp_pdf(grid, 1),ax=axes[0],label='pdf',linestyle='--')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=exp_cdf(grid, 1),ax=axes[1],label='cdf',linestyle='--')
axes[1].set_title('Sample ECDF')
plt.show()
```

```python
r_values = [2, 5, 10]
p_values = [0.3, 0.5, 0.7]
k = np.arange(0, 30)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for r in r_values:
    for p in p_values:
        cdf = sp.stats.nbinom.cdf(k, r, p)
        axes[1].plot(k, cdf, label=f"r={r}, p={p}")

axes[0].set_title("Negative Binomial PMF for various r, p")
axes[0].set_xlabel("k")
axes[0].set_ylabel("Cumulative Probability")
axes[0].legend()

for r in r_values:
    for p in p_values:
        pmf = sp.stats.nbinom.pmf(k, r, p)
        axes[0].plot(k, pmf, label=f"r={r}, p={p}")

axes[1].set_title("Negative Binomial CDF for various r, p")
axes[1].set_xlabel("k (number of failures before r successes)")
axes[1].set_ylabel("Probability")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Sample with 1000 draws, plot KDE/ECDF compared to theoretical pdf/cdf

grid = np.arange(0, 30)
sample = sp.stats.nbinom.rvs(5, 0.5, size=1000, random_state=10)

pdf = sp.stats.nbinom.pmf(grid, r, p)
cdf = sp.stats.nbinom.cdf(grid, r, p)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
sns.lineplot(x=grid,y=pdf,ax=axes[0],label='pdf',linestyle='--')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=cdf,ax=axes[1],label='cdf',linestyle='--')
axes[1].set_title('Sample ECDF')
plt.show()
```

- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

#### Answer
- Support: the support for the categorical pmf is $k = 1, 2, 3, ..., K$
- Plotting density and distribution for variety of values:

```python
p = np.array([0.1, 0.2, 0.3, 0.25, 0.15])  # probabilities for 5 categories
K = len(p)
categories = np.arange(1, K + 1)

categorical_dist = sp.stats.rv_discrete(name='categorical', values=(categories, p))

pmf = categorical_dist.pmf(categories)
cdf = categorical_dist.cdf(categories)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].stem(categories, pmf, basefmt=" ")
# axes[0].plot(categories, pmf)
axes[0].set_title("Categorical PMF")
axes[0].set_xlabel("Category (k)")
axes[0].set_ylabel("Probability")
axes[0].set_xticks(categories)

axes[1].step(categories, cdf, where='post')
# axes[1].plot(categories, cdf)
axes[1].set_title("Categorical CDF")
axes[1].set_xlabel("Category (k)")
axes[1].set_ylabel("Cumulative Probability")
axes[1].set_xticks(categories)

plt.tight_layout()
plt.show()

```

```python
# Sample with 1000 draws, plot KDE/ECDF compared to theoretical pdf/cdf

sample = categorical_dist.rvs(size=1000, random_state=42)
grid = np.arange(1, K + 1)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# ECDF vs CDF
sns.ecdfplot(x=sample, ax=axes[0], label='Empirical CDF')
sns.lineplot(x=grid, y=cdf, ax=axes[0], label='Theoretical CDF', linestyle='--')
axes[0].set_title("Categorical ECDF vs Theoretical CDF")
axes[0].set_xlabel("Category (k)")
axes[0].legend()

# Histogram vs PMF
# sns.histplot(sample, bins=np.arange(0.5, K + 1.5, 1), stat='probability', ax=axes[1], label='Sample Histogram')
sns.kdeplot(x=sample, ax=axes[1])
sns.lineplot(x=grid, y=pmf, ax=axes[1], label='Theoretical PMF', linestyle='--', marker='o')
axes[1].set_title("Categorical Sample vs Theoretical PMF")
axes[1].set_xlabel("Category (k)")
axes[1].legend()

plt.tight_layout()
plt.show()
```
